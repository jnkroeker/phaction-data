```python
#Create new dataframe with only columns I care about
cols = [ 'upload_id', 'type', 'distance', 'moving_time',   
         'elapsed_time', 'average_speed', 'max_speed',
         'average_heartrate', 'max_heartrate',
         'total_elevation_gain', 'start_date_local',
         'map'
       ]
activities = wrkouts[cols]
```

```python
activities.head()
```

```python
# Break start_date_local into start time and date
# This makes is easier to match up with SugarWOD workouts 
activities['start_date_local'] = pd.to_datetime(activities['start_date_local'])
activities['start_time'] = activities['start_date_local'].dt.time
activities['start_date_local'] = activities['start_date_local'].dt.date
```

```python
activities.head()
```

```python
activities['type'].value_counts()
```

```python
# Create a table for Crossfit and Ride

crossfit = activities.loc[activities['type'] == 'Crossfit']
cycling  = activities.loc[activities['type'] == 'Ride']
```

```python
# On average, do I reach a higher maximum heartrate during Crossfit workouts or on bike rides?

cf_max_mean = round(crossfit['max_heartrate'].mean(), 2)
br_max_mean = round(cycling['max_heartrate'].mean(), 2)
print("Mean Crossfit Max Heartrate: " + str(cf_max_mean) + " | Mean Cycling Max Heartrate: " + str(br_max_mean))
```

```python
# On average, do I maintain a higher average heartrate during bike rides than Crossfit workouts?

cf_avg_mean = round(crossfit['average_heartrate'].mean(), 2)
br_avg_mean = round(cycling['average_heartrate'].mean(), 2)
print("Mean Crossfit Average Heartrate: " + str(cf_avg_mean) + " | Mean Cycling Average Heartrate: " + str(br_avg_mean))
```
